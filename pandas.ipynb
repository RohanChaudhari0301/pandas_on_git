```python
dict2 = {
    "name":['rohan','batman','rohan'],
    "toy":[np.nan,'Batmobile','Codes'],
    "born":[pd.NaT,pd.Timestamp("1940-04-25"),pd.NaT]
}

df = pd.DataFrame(dict2)
df.head()
```

```python
df.dropna() # will drop all the rows where there is atleast one Nan value
```

```python
df.dropna(how='all') # will drop all the rows where all the columns have nan values in that row
```

```python
df.dropna(how='all',axis = 1) # will drop all the columns where all the rows have nan values in that row
```

# drop duplicated

```python
df.drop_duplicates(subset=['name'])
```

```python
df.drop_duplicates(subset=['name'],keep='first')
```

```python
df.drop_duplicates(subset=['name'],keep='last')
```

```python
df.drop_duplicates(subset=['name'],keep=False)
```
